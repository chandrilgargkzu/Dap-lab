```python
#Q3
import numpy as np

def get_matrix():
    return np.array([list(map(float, input().split())) for _ in range(2)])

print("Matrix A:")
A = get_matrix()

print("Matrix B:")
B = get_matrix()

print(A+B)
print(A*B)
print(A@B)
```

```python
#Q4
import numpy as np

A = np.array([list(map(float, input("Enter 2x2 matrix row: ").split())) for _ in range(2)])

print(A.T)

try:
    print(np.linalg.inv(A))
except:
    print("❌ Inverse not possible")
```

```python
#Q8
import numpy as np

a = np.array(list(map(float, input("a: ").split())))
b = np.array(list(map(float, input("b: ").split())))

print(np.dot(a,b))

if len(a)==3 and len(b)==3:
    print(np.cross(a,b))
else:
    print("Cross product needs 3D vectors")
```

```python
#Q9
import numpy as np

a = np.array(list(map(float, input("Enter elements: ").split())))

r = int(input("Rows: "))
c = int(input("Cols: "))

if len(a) == r*c:
    mat = a.reshape(r,c)
    print(mat)
    print(mat.flatten())
else:
    print("Cannot reshape")
```

```python
#Q10
import numpy as np, time

n = int(input("Size: "))

lst = list(range(n))
arr = np.array(lst)

t1 = time.time()
sum(lst)
print("List:", time.time()-t1)

t2 = time.time()
np.sum(arr)
print("NumPy:", time.time()-t2)
```

```python
#Q11
from statistics import mode

data = list(map(float, input("Enter data: ").split()))

mean = sum(data)/len(data)
data.sort()
median = data[len(data)//2]

try:
    m = mode(data)
except:
    m = "No unique mode"

print(mean, median, m)
```

```python
#Q16
import numpy as np, matplotlib.pyplot as plt

mu = float(input("Mean: "))
sigma = float(input("Std: "))

x = np.linspace(-5,5,100)
y = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-(x-mu)**2/(2*sigma**2))

plt.plot(x,y)
plt.show()
```

```python
#Q19
from scipy import stats

data = list(map(float, input("Sample: ").split()))
mean = float(input("Population mean: "))

t,p = stats.ttest_1samp(data, mean)

print(t,p)
```

```python
#Q20
from scipy.stats import chi2_contingency

print("Enter 2x2 table:")
table = [list(map(float, input().split())) for _ in range(2)]

chi2,p,_,_ = chi2_contingency(table)

print(chi2,p)
```
